```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_sepia(image):
    """
    Applied a sepia filter to the input image.

    Parameters:
        image (numpy.ndarray): The input image in RGB format.

    Returns:
        numpy.ndarray: The image with sepia filter applied.
    """
    # Defined the sepia filter matrix
    sepia_filter = np.array([[0.272, 0.534, 0.131],
                             [0.349, 0.686, 0.168],
                             [0.393, 0.769, 0.189]])
    # Applied the sepia filter to the image
    sepia_image = cv2.transform(image, sepia_filter)
    # Clipped the pixel values to be in the valid range
    sepia_image = np.clip(sepia_image, 0, 255)
    return sepia_image

def main():
    # Loaded the image
    image = cv2.imread('Untitled.jpeg')
    # Converted the image from BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Converted the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Applied the sepia filter
    sepia_image = apply_sepia(image_rgb)

    # Detected edges using Canny edge detector
    edges = cv2.Canny(gray_image, 100, 200)

    # Plotted the original image and processed images
    plt.figure(figsize=(12, 8))

    # Original Image
    plt.subplot(2, 2, 1)
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.title('Original Image')

    # Grayscale Image
    plt.subplot(2, 2, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.axis('off')
    plt.title('Grayscale Image')

    # Sepia Image
    plt.subplot(2, 2, 3)
    plt.imshow(sepia_image)
    plt.axis('off')
    plt.title('Sepia Image')

    # Edge Detection Image
    plt.subplot(2, 2, 4)
    plt.imshow(edges, cmap='gray')
    plt.axis('off')
    plt.title('Edges')

    plt.show()

if __name__ == "__main__":
    main()
```
